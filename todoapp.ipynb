```python
items = []

def add_item(item):
    items.append(item)
    
def show_item(items):
    sno = 0
    for item in items:
        sno+=1
        print(str(sno)+" "+item)

def remove_all(items):
    items.clear()
    print("Everything Removed.Nothing TODO")
    
def remove_one(items):
    show_item(items)
    itemNo = int(input("Which item to remove?\n"))
    indexNo = itemNo -1
    del items[indexNo]
    print("**Item Removed**")
    show_item(items)
    
while(True):
    print("\t\t\t****TODO APP****")
    item = input("What to do?\n 1)Add Item\n 2)Remove Item\n q to quit:\n")
    if item == "q":
        break
    else:
        itemss = int(item)
        if itemss == 1:
            addItem = input("What TODO?")
            print("\t**TODO LIST**")
            add_item(addItem)
            show_item(items)
        elif itemss == 2:
            uinput = int(input(" 1)Remove All\n 2)Remove One\n"))
            if uinput == 1:
                remove_all(items)
            else:
                remove_one(items)
```
